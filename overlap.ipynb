```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
file_path = "data/overlap_df.csv"
df = pd.read_csv(file_path, index_col=0)

# Extract keyword names
keywords = df.index.tolist()
```

```python
print(keywords)
```

```python

overlap_matrix = df.to_numpy()
overlap_matrix = overlap_matrix / 50
print(overlap_matrix.shape)
```

```python
S = np.sum(overlap_matrix, axis=1)
print(S)
```

```python
# get rows of a numpy matrix
rows, cols = overlap_matrix.shape
A = np.zeros(rows)
for i in range(cols):
    sum = 0
    for j in range(rows):
        sum = sum+ overlap_matrix[i,j]*S[j]
    sum = sum/rows
    A[i]=sum
print(A)
```

```python
# Normalize Scores
min_A, max_A = np.min(A), np.max(A)
N_i = (A - min_A) / (max_A - min_A) * 5

N_i = np.trunc(N_i).astype(int)  #[2. 2. 1. 3. 3. 4. 3. 0. 3. 2. 3. 5. 3. 4. 5. 3. 2. 4. 3. 2. 2. 2. 2. 2. 2. 4. 5. 0. 4.]
N_i= - N_i +6
print(N_i)
```

```python
# df_result = df_result.sort_values(by="Final Search Weight (W)", ascending=False)

result_array = np.vstack((keywords, N_i))
result_array = np.vstack((keywords, S))
print(result_array.shape)

df_result = pd.DataFrame(result_array.T)
from IPython.display import display
display(df_result)
```
